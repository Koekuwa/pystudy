```python
#高速フーリエ
```

```python
import datetime
import numpy as np
import scipy as sp
import scipy.fftpack
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = pd.read_csv("weather.csv", na_values=(-9999), parse_dates=["DATE"])
df = df0[df0["DATE"]>="19940101"]
df_avg = df.dropna().groupby("DATE").mean()
date = df_avg.index.to_datetime()
temp = (df_avg["TMAX"] + df_avg["TMIN"]) / 20.
N = len(temp)
plt.plot_date(date, temp, "-", lw=.5)
plt.ylim(-10, 40)
plt.xlabel("Date")
plt.ylabel("Mean temperature")
```

```python
temp_fft = sp.fftpack.fft(temp)
temp_psd = np.abs(temp_fft) ** 2
fftfreq = sp.fftpack.fftfreq(len(temp_psd), 1./365)
i = fftfreq>0
plt.plot(fftfreq[i], 10*np.log10(temp_psd[i]))
plt.xlim(0, 5)
plt.xlabel("Frequency (1/year)")
plt.ylabel("PSD (dB)")
```

```python
temp_fft_bis = temp_fft.copy()
temp_fft_bis[np.abs(fftfreq) > 1.1] = 0
temp_slow = np.real(sp.fftpack.ifft(temp_fft_bis))
plt.plot_date(date, temp, "-", lw=.5)
plt.plot_date(date, temp_slow, "-")
plt.xlim(datetime.date(1994, 1, 1), datetime.date(2000, 1, 1))
plt.ylim(-10, 40)
plt.xlabel("Date")
plt.ylabel("Mean temperature")
```

```python
#線形フィルタ
```

```python
import numpy as np
import scipy as sp
import scipy.signal as sg
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nasdaq_df = pd.read_csv("nasdaq.csv")
nasdaq_df.head()
```

```python
date = pd.to_datetime(nasdaq_df["Date"])
nasdaq = nasdaq_df["Close"]
plt.plot_date(date, nasdaq, "-")
```

```python
h = sg.get_window("triang", 60)
fil = sg.convolve(nasdaq, h/h.sum())
plt.plot_date(date, nasdaq, "-", lw=1)
plt.plot_date(date, fil[:len(nasdaq-1)], "-")
```

```python
plt.plot_date(date, nasdaq, "-", lw=1)
b, a = sg.butter(4, 2./365)
plt.plot_date(date, sg.filtfilt(b, a, nasdaq), "-")
```

```python
plt.plot_date(date, nasdaq, "-", lw=1)
b, a = sg.butter(4, 2*5./365, btype="high")
plt.plot_date(date, sg.filtfilt(b, a, nasdaq), "-")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
files = [file for file in os.listdir("data/") if file.startswith("yob")]
years = np.array(sorted([int(file[3:7]) for file in files]))
data = {year: pd.read_csv("data/yob{y:d}.txt".format(y=year), index_col=0, header=None, 
                          names=["First name", "Gender", "Number"]) for year in years}
data[2012].head()
```

```python
def get_value(name, gender, year):
    try:
        return data[year] \
               [data[year]["Gender"] == gender] \
               ["Number"][name]
    except KeyError:
        return 0
                             
def get_evolution(name, gender):
    return np.array([get_value(name, gender, year) for year in years])

def autocorr(x):
    result = np.correlate(x, x, mode="full")
    return result[result.size/2:]

def autocorr_name(name, gender):
    x = get_evolution(name, gender)
    z = autocorr(x)
    plt.subplot(121)
    plt.plot(years, x, "o", label=name)
    plt.title("Baby names")
    plt.subplot(122)
    plt.plot(z/float(z.max()), "-", label=name)
    plt.legend()
    plt.title("Autocorrelation")
```

```python
autocorr_name("Olivia", "F")
autocorr_name("Maria", "F")
```
