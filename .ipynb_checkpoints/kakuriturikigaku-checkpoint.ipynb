```python
#マルコフ連鎖
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
a = 0.5 / N
b = 0.5 / N
nsteps = 1000
x = np.zeros(nsteps)
x[0] = 25

for t in range(nsteps - 1):
    if 0 < x[t] < N-1:
        birth = np.random.rand() <= a*x[t]
        death = np.random.rand() <= b*x[t]
        x[t+1] = x[t] + 1*birth - 1*death
    else:
        x[t+1] = x[t]
plt.plot(x)
```

```python
ntrials = 100
x = np.random.randint(size=ntrials, low=0, high=N)

def simulate(x, nsteps):
    for _ in range(nsteps - 1):
        upd = (0 < x) & (x < N-1)
        birth = 1*(np.random.rand(ntrials) <= a*x)
        death = 1*(np.random.rand(ntrials) <= b*x)
        x[upd] += birth[upd] - death[upd]
        
bins = np.linspace(0, N, 25)
nsteps_list = [10, 1000, 10000]
for i, nsteps in enumerate(nsteps_list):
    plt.subplot(1, len(nsteps_list), i+1)
    simulate(x, nsteps)
    plt.hist(x, bins=bins)
    plt.xlabel("Population size")
    if i == 0:
        plt.ylabel("Histogram")
    plt.title("{0:d} time steps.".format(nsteps))
```

```python
#ポアソン過程
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rate = 20.
dt = .001
n = int(1./dt)

x = np.zeros(n)
x[np.random.rand(n) <= rate*dt] = 1
plt.vlines(np.nonzero(x)[0], 0, 1)
plt.xticks([]); plt.yticks([])
```

```python
plt.plot(np.linspace(0., 1., n), np.cumsum(x))
plt.xlabel("Time")
plt.ylabel("Counting progress")
```

```python
y = np.cumsum(np.random.exponential(1./rate, size=int(rate)))
plt.vlines(y, 0, 1)
plt.xticks([]); plt.yticks([])
```

```python
#ブラウン運動
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 5000

x = np.cumsum(np.random.randn(n))
y = np.cumsum(np.random.randn(n))

k = 10
x2 = np.interp(np.arange(n*k), np.arange(n)*k, x)
y2 = np.interp(np.arange(n*k), np.arange(n)*k, y)
plt.scatter(x2, y2, c=range(n*k), linewidth=0, marker="o", s=3, cmap=plt.cm.jet)
plt.axis("equal")
plt.xticks([]); plt.yticks([])
```

```python
#確率微分方程式
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sigma = 1.
mu = 10.0
tau = 0.05
dt = 0.001
T = 1.0
n = int(T/dt)
t = np.linspace(0., T, n)

sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt = np.sqrt(dt)
x = np.zeros(n)
for i in range(n-1):
    x[i+1] = x[i] + dt*(-(x[i]-mu)/tau) + sigma_bis * sqrtdt * np.random.randn()
plt.plot(t, x)
```

```python
ntrials = 10000
X = np.zeros(ntrials)
bins = np.linspace(-2., 14., 50);
for i in range(n):
    X += dt*(-(X-mu)/tau) + sigma_bis * sqrtdt * np.random.randn(ntrials)
    if i in (5, 50, 900):
        hist, _ = np.histogram(X, bins=bins)
        plt.plot((bins[1:]+bins[:-1])/2, hist, label="t={0:.2f}".format(i*dt))
    plt.legend()
```
