```python
import numpy as np
import scipy as sp
import scipy .optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: 1-np.sin(x)/x
x = np.linspace(-20., 20., 1000)
y = f(x)
plt.plot(x, y)
```

```python
x0 = 3
xmin = opt.minimize(f, x0).x
plt.plot(x, y)
plt.scatter(x0, f(x0), marker="o", s=300)
plt.scatter(xmin, f(xmin), marker="v", s=300)
plt.xlim(-20,20)
```

```python
x0 = 10
xmin = opt.minimize(f, x0).x
plt.plot(x, y)
plt.scatter(x0, f(x0), marker="o", s=300)
plt.scatter(xmin, f(xmin), marker="v", s=300)
plt.xlim(-20,20)
```

```python
x0 = 10
xmin = opt.minimize(f, x0, method="Anneal").x
plt.plot(x, y)
plt.scatter(x0, f(x0), marker="o", s=300)
plt.scatter(xmin, f(xmin), marker="v", s=300)
plt.xlim(-20,20)
```

```python
def g(X):
    x, y = X
    return(np.sin(3*np.pi*x)**2 + (x-1)**2 * (1+np.sin(3*np.pi*x)**2) + (y-1)**2 * (1+np.sin(3*np.pi*x)**2))

n = 200
k = 10
X, Y =np.mgrid[-k:k:n*1j, -k:k:n*1j]
Z = g(np.vstack((X.ravel(), Y.ravel()))).reshape(n, n)
plt.imshow(np.log(Z), cmap=plt.cm.hot_r)
plt.xticks([]); plt.yticks([])
```

```python
x0 ,y0 = opt.minimize(g, (8, 3)).x
plt.imshow(np.log(Z), cmap=plt.cm.hot_r, extent=(-k, k, -k, k), origin=0)
plt.scatter([x0], [y0], c=["r"], s=100)
plt.xticks([]); plt.yticks([])
```

```python
import numpy as np
import scipy .optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x, a, b, c, d):
    return a/(1 + np.exp(-c * (x-d))) +b

a, c = np.random.exponential(size=2)
b, d = np.random.randn(2)
n = 100
x = np.linspace(-10., 10., n)
y_model = f(x, a, b, c, d)
y = y_model + a * .2 * np.random.randn(n)
plt.plot(x, y_model, "--k")
plt.plot(x, y, "o")
```

```python
(a_, b_, c_, d_), _ =opt.curve_fit(f, x, y, (a, b, c, d))
y_fit = f(x, a_, b_, c_, d_)
plt.plot(x, y_model, "--k")
plt.plot(x, y, "o")
plt.plot(x, y_fit, "-")
```
