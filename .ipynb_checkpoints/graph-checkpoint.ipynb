```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 10
adj = [(i, (i+1)%n) for i in range(n)]
adj += [(i, (i+2)%n) for i in range(n)]
g = nx.Graph(adj)
print(g.nodes())
print(g.edges())
print(nx.adjacency_matrix(g))
```

```python
nx.draw_circular(g)
```

```python
g.add_node(n, color="#fcff00")
for i in range(n):
    g.add_edge(i, n)
t = np.linspace(0., 2*np.pi, n)
pos = np.zeros((n+1, 2))
pos[:n, 0] = np.cos(t)
pos[:n, 1] = np.sin(t)
color = [g.node[i].get("color", "#88b0f3") for i in range(n+1)]
nx.draw_networkx(g, pos=pos, node_color=color)
plt.axis("off")
```

```python
nx.draw_spectral(g, node_color=color)
plt.axis("off")
```

```python
import math
import json
import twitter
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
(CONSUMER_KEY, CONSUMER_SECRET, OAUTH_TOKEN, OAUTH_TOKEN_SECRET) = open("twitter.txt", "r").read().splitlines()
auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET, CONSUMER_KEY, CONSUMER_SECRET)
tw = twitter.Twitter(auth=auth)

me = tw.account.verify_credentials()
myid = me["id"]

def get_followers_ids(uid=None):
    return tw.followers.ids(user_id=uid)["ids"]

my_followers_ids = get_followers_ids()

def get_users_info(users_ids, mx=500):
    n = min(mx, len(users_ids))
    users = [tw.users.lookup(user_id=users_ids[100*i:100*(i+1)]) for i in range(int(math.ceil(n/100.)))]
    users = [item for sublist in users for item in sublist]
    return {user["id"]: user for user in users}

users_info = get_users_info(my_followers_ids)
with open("my_followers.json", "w") as f:
    json.dump(users_info, f, indent=1)
```

```python
adjacency = {myid: my_followers_ids}
my_followers_nagoya = [user for user in users_info.values() if "名古屋" in user["description"].lower()]
my_followers_nagoya_best = sorted(my_followers_nagoya, key=lambda u: u["followers_count"])[::-1][:10]

tw.application.rate_limit_status(resources="followers")["resources"]["followers"]["/followers/ids"]
```

```python
for user in my_followers_nagoya_best:
    adjacency[user["id"]] = list(set(get_followers_ids(user["id"])).intersection(my_followers_ids))
    
g = nx.Graph(adjacency)
g = g.subgraph(users_info.keys())
with open("my_graph.json", "w") as f:
    json.dump(nx.to_dict_of_lists(g), f, indent=1)
    
g.remove_nodes_from([k for k, d in g.degree().items() if d <= 1])
g.remove_nodes_from([myid])
len(g.nodes()), len(g.edges())
```

```python
deg = g.degree()
for user in users_info.values():
    fc = user["followers_count"]
    sc = user["statuses_count"]
    user["python"] = "python" in user["description"].lower()
    user["node_size"] = math.sqrt(10 + 50 * fc)
    user["node_color"] = 10 * math.sqrt(1 + sc)
    user["label"] = user["screen_name"] if fc > 500 else ""

node_size = [users_info[uid]["node_size"] for uid in g.nodes()]
node_color = [users_info[uid]["node_color"] for uid in g.nodes()]
labels = {uid: users_info[uid]["label"] for uid in g.nodes()}
nx.draw(g, cmap=plt.cm.OrRd, alpha=.8, node_size=node_size, node_color=node_color, labels=labels, font_size=8, width=.1)
```

```python
#画像連結成分処理
```

```python
import itertools
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as col
%matplotlib inline
```

```python
n = 10
img = np.random.randint(size=(n, n), low=0, high=3)
g = nx.grid_2d_graph(n, n)

def show_image(img, **kwargs):
    plt.imshow(img, origin="lower", interpolation="none", **kwargs)
    plt.axis("off")

def show_graph(g, **kwargs):
    nx.draw(g, pos={(i, j): (j, i) for (i, j) in g.nodes()}, node_color=[img[i, j] for (i, j) in g.nodes()],
            linewidth=1, edge_solor="w", with_labels=False, node_size=30, **kwargs)

cmap = plt.cm.Blues
```

```python
show_image(img, cmap=cmap, vmin=-1)
show_graph(g, cmap=cmap, vmin=-1)
```

```python
g2 = g.subgraph(zip(*np.nonzero(img==2)))
show_image(img, cmap=cmap, vmin=-1)
show_graph(g2, cmap=cmap, vmin=-1)
```

```python
components = [np.array(comp) for comp in nx.connected_components(g2) if len(comp) >= 3]
len(components)
print(components)
```

```python
img_bis = img.copy()
for i, comp in enumerate(components):
    img_bis[comp[:,0], comp[:,1]] = i + 3
colors = [cmap(.5), cmap(.75), cmap(1.), "#f4f235", "f4a535", "f44b35"]
cmap2 = col.ListedColormap(colors, "indexed")
show_image(img_bis, cmap=cmap2)
```
