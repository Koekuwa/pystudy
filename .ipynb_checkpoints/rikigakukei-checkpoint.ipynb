```python
#カオス分岐図
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic(r, x):
    return r*x*(1-x)
```

```python
n = 10000
r = np.linspace(2.5, 4.0, n)
iterations = 1000
last = 100
x = 1e-5*np.ones(n)
lyapunov = np.zeros(n)
plt.subplot(211)
for i in range(iterations):
    x = logistic(r, x)
    lyapunov += np.log(abs(r-2*r*x))
    if i >= (iterations - last):
        plt.plot(r, x, ",k", alpha=.02)
plt.xlim(2.5, 4)
plt.title("Bifurcation diagram")

plt.subplot(212)
plt.plot(r[lyapunov<0], lyapunov[lyapunov<0] / iterations, ",r", alpha=0.1)
plt.plot(r[lyapunov>=0], lyapunov[lyapunov>=0] / iterations, ",b", alpha=0.25)
plt.xlim(2.5, 4)
plt.ylim(-2, 1)
plt.title("Lyapunov exponent")
```

```python
#セルオートマトン
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
u = np.array([[4],[2],[1]])
def step(x, rule_binary):
    y = np.vstack((np.roll(x, 1), x, np.roll(x, -1))).astype(np.int8)
    z = np.sum(y * u, axis=0).astype(np.int8)
    return rule_binary[7-z]

def generate(rule, size=80, steps=80):
    rule_binary = np.array([int(x) for x in np.binary_repr(rule, 8)], dtype=np.int8)
    x = np.zeros((steps, size), dtype=np.int8)
    x[0,:] = np.random.rand(size) < .5
    for i in range(steps-1):
        x[i+1,:] = step(x[i,:], rule_binary)
    return x
```

```python
rules = [3, 18, 30, 90, 106, 110, 158, 154, 184]
for i, rule in enumerate(rules):
    x = generate(rule)
    plt.subplot(331+i)
    plt.imshow(x, interpolation="none", cmap=plt.cm.binary)
    plt.xticks([]); plt.yticks([])
    plt.title(str(rule))
```

```python
#常微分方程式
```

```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = 1.
k = 1.
g = 9.81
v0 = np.zeros(4)
v0[2] = 4.
v0[3] = 10.

def f(v, t0, k):
    u, udot = v[:2], v[2:]
    udotdot = -k/m * udot
    udotdot[1] -= g
    return np.r_[udot, udotdot]

t = np.linspace(0., 3., 30)
for k in np.linspace(0., 1., 5):
    v = spi.odeint(f, v0, t, args=(k,))
    plt.plot(v[:,0], v[:,1], "o-", mew=1, ms=8, mec="w", label="k={0:.1f}".format(k))
plt.legend()
plt.xlim(0, 12)
plt.ylim(0, 6)
```

```python
#偏微分方程式
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 2.8e-4
b = 5e-3
tau = .1
k = -.005
size = 80
dx = 2. / size
T = 10.0
dt = .9 * dx**2 / 2
n = int(T / dt)

U = np.random.rand(size, size)
V = np.random.rand(size, size)

def laplacian(Z):
    Ztop = Z[0:-2, 1:-1]
    Zleft = Z[1:-1, 0:-2]
    Zbottom = Z[2:, 1:-1]
    Zright = Z[1:-1, 2:]
    Zcenter = Z[1:-1, 1:-1]
    return (Ztop + Zleft + Zbottom + Zright - 4 * Zcenter) / dx**2

for i in range(n):
    deltaU = laplacian(U)
    deltaV = laplacian(V)
    Uc = U[1:-1, 1:-1]
    Vc = V[1:-1, 1:-1]
    U[1:-1, 1:-1], V[1:-1, 1:-1] = (Uc + dt * (a*deltaU + Uc - Uc**3 - Vc +k), Vc + dt * (b*deltaV + Uc - Vc) / tau)
    for Z in (U, V):
        Z[0,:] = Z[1,:]
        Z[-1,:] = Z[-2,:]
        Z[:,0] = Z[:,1]
        Z[:,-1] = Z[:,-2]
plt.imshow(U, cmap=plt.cm.copper, extent=[-1, 1, -1, 1])
```
